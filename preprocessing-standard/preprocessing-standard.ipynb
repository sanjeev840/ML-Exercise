```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
data = {
    'age': [25, 30, 35, 40],
    'income': [50000, 60000, 70000, 80000],
    'city': ['New York', 'Los Angeles', 'Chicago', 'New York']
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

# Standard scalar - converting the data to mean=0 and SD = 1

```python
scaler = StandardScaler()
scaler.fit(df[['age', 'income']])
```

```python
scaler.mean_
```

```python
scaler.var_
```

```python
scaled_income = scaler.transform(df[['age', 'income']])
```

```python
pd.DataFrame(scaled_income, columns = scaler.get_feature_names_out())
```

# Min Max Scaler

```python
minmax = MinMaxScaler()
minmax.fit(df[['age', 'income']])
```

```python
minmax_income = minmax.transform(df[['age', 'income']])
```

```python
pd.DataFrame(minmax_income, columns = minmax.get_feature_names_out())
```

# One Hot Encoder

```python
transformer=ColumnTransformer(transformers =[('onehot', OneHotEncoder(handle_unknown='ignore'),['city'])],remainder='passthrough')
```

```python
ohe_output = transformer.fit_transform(df)
```
